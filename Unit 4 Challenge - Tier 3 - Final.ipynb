**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Created a DataFrame with yearly average prices for each Borough, named it dfg based on the suggestion:
dfg = df.groupby(by = ['London_Borough', 'Year']).mean()
dfg.head(10)
```

```python
# Reset the index for dfg
dfg = dfg.reset_index()
dfg.head()
```

```python
dfg[dfg['London_Borough']=='Barnet'].head()
```

```python
# Trying to figure out a path to get to the 'Average_price' value in the dfg DataFrame
path1 = dfg[dfg['London_Borough'] == 'Barnet']
path1[path1['Year'] == 2000]['Average_price']
```

```python
# Let's create the function
def create_price_ratio(borough, year1, year2):
    """Calculate the price ratio based on average housing price for two years"""
    path = dfg[dfg['London_Borough'] == borough]
    price_y1 = float(path[path['Year'] == year1]['Average_price'])
    price_y2 = float(path[path['Year'] == year2]['Average_price'])
    price_ratio = price_y1/price_y2
    return price_ratio
```

```python
# Testing the function for 'Barnet' borough
create_price_ratio('Barnet', 2000, 2020)
```

```python
# Will have to create a new DataFrame with the final values for Boroughs and price_ratios, based on a for loop
# First created any empty dictionary to get save values in for the for loop
final_dict = {}
```

```python
# Figuring out how to build the for loop structure
for x in dfg['London_Borough'].unique():
    print(x)
```

```python
for x in dfg['London_Borough'].unique():
    final_dict[x] = create_price_ratio(x, 2000, 2020)

final_dict
```

```python
# Converting this dictionary to DataFrame
final_df = pd.DataFrame(final_dict.items(), columns=['London_Borough', 'Price_ratio'])
final_df
```

```python
# Let's sort the DataFrame to get the price ratio values in descending order
final_df_sorted = final_df.sort_values(by = 'Price_ratio', ascending = False)
final_df_sorted
```

```python
# It can be seen that Bromley has the highest average housing price increase amongst the City of London Boroughs, 
# but for practical purposes Hammersmith & Fulham and Harrow Boroughs could be said to have the same increase
# We can focus top 10 boroughs and represent the data graphically
final_df_top10 = final_df_sorted.head(10)

final_df_top10.plot(kind = 'bar', x = 'London_Borough')
plt.xlabel('London Borough', fontsize=10)
plt.ylabel('Price Ratio', fontsize=10)
plt.show()
```
